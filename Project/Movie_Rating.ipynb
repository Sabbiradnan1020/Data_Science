```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import sqlite3


df=pd.read_csv('imdb.csv')

# print(df.shape)
# print(df.columns)
# print(df.head())
print(df.isnull().sum())


df['Rating_from_10'].fillna(df['Rating_from_10'].mean(),inplace=True)

#genar_wise average rating

genre_avg_rating = df.groupby('Genre')['Rating_from_10'].mean().sort_values(ascending=False).head(10)
short_labels = {
    'Adventure, Western': 'Adv/West',
    'Crime, Drama, Fantasy': 'Crime/Drama/Fan',
    'Drama, Family, Fantasy': 'Drama/Fam/Fan',
    'Animation, Drama, War': 'Anim/Drama/War',
    'Drama, Mystery, War': 'Drama/Myst/War',
    'Action, Sci-Fi': 'Act/Sci-Fi',
    'Adventure, Drama, Sci-Fi': 'Adv/Drama/SF',
    'Comedy, Musical, Romance': 'Com/Mus/Rom',
    'Crime, Sci-Fi': 'Crime/SF',
    'Action, Adventure': 'Act/Adv'
}

# Apply label mapping
labels = [short_labels.get(g, g) for g in genre_avg_rating.index]

sns.barplot(x=labels, y=genre_avg_rating.values)
plt.xticks(rotation=90)
plt.title('Average Rating by Genre')
plt.show()

sns.histplot(df['Rating_from_10'], bins=5, kde=True)
plt.title('Distribution of Ratings')
plt.show()


```
