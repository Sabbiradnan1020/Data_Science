```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import sqlite3

df=pd.read_csv('COVID-19.csv')
# # print(df.shape)
# print(df.columns)
# print(df.head())
# # print(df.isnull().sum())

# Data Cleaning - Remove Duplicates
df.fillna(0,inplace=True)
df.drop_duplicates(subset=['Country/Region'], inplace=True)
#Basic statistics


print('Mean of Cases:',df['Confirmed'].mean())
print('median of Cases:',df['Confirmed'].median())
print('Standard Deviation of Cases:',df['Confirmed'].std())

#EDA
print(df.describe())
print(df.info())

#Data Visualization 

top10_countries=df.sort_values(by='Confirmed', ascending=False).head(10)
plt.figure(figsize=(10, 6))
sns.barplot(x=top10_countries['Country/Region'],y=top10_countries['Confirmed'], color='skyblue')
plt.title('Confirmed Cases by Country/Region')
plt.xlabel('Country/Region')
plt.ylabel('Confirmed Cases')
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()

sns.scatterplot(x=df['New cases'],y=df['New deaths'],color='red',alpha=0.6)
plt.title('New Cases vs New Deaths')
plt.xlabel('New Cases')
plt.ylabel('New Deaths')
plt.show()







```
