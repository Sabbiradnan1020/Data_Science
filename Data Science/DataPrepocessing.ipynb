```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.preprocessing import LabelEncoder,MinMaxScaler,StandardScaler
data = {
    'Name': ['Alice', 'Bob', 'Charle', 'David', 'Ravindra', 'Maxwell', 'Helen', 'Frank'],
    'Age': [23, 45, 60, 23, np.nan, 55, 12, 34],  
    'Salary': [5000, 60000, 90000, 4333, 100000, 9000, 900000, 12000000],
    'Gender': ['Female', 'Male', 'Male', 'Male', 'Male', 'Male', 'Female', 'Male'],
    'City': ['London', 'Bangladesh', 'France', 'New York', 'Chicago', 'Chicago', 'Osaka', 'Houston']  
}

df = pd.DataFrame(data)

# df['Age'].fillna(df['Age'].mean(),inplace=True)    
# print(df.isnull().sum())
# print(df)       

sns.boxplot(x=df['Salary'])
plt.title("Salary Box Plot (Before Removing Outliers)")
plt.show()
Q1=df['Salary'].quantile(0.25)
Q3=df['Salary'].quantile(0.75)
IQR=Q3-Q1

df_cleand=df[(df['Salary']>=Q1-1.5*IQR) & (df['Salary']<=Q3+1.5*IQR)]
print('After filter outline')
print(df_cleand)

sns.boxplot(X=df_cleand['Salary'])
# plt.title("Salary Box Plot (After Removing Outliers)")
# plt.show()
```
