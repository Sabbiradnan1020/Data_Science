```python
import matplotlib.pyplot as plt
x=[1,2,3,4,5,6]
y=[10,20,30,40,50,60]
plt.plot(x,y,marker='o',linestyle='-',color='c')
plt.xlabel("X Axis")
plt.ylabel("Y Axis")
plt.title("Simple line chart")
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
data={
    'names': ['Atif','Mahin','Sifat','Hamja','Himal','Rupa','Suchi','Naim','Soikot','Foysal'],
'ages':[ 20,30,40,50,80,13,15,74,22,20]
}
df=pd.DataFrame(data)
plt.bar(df['names'],df['ages'],color='red')
plt.xlabel("Name")
plt.ylabel("Age")
plt.title("Age Distribution")
plt.xticks(rotation=45)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

data=np.random.randn(1000)
plt.hist(data,bins=20,color='green',edgecolor='red')
plt.xlabel("value range")
plt.ylabel('Frequency')
plt.title("Histogram exmale")
plt.show()
```
